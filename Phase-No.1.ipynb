```python
import pandas as p
import matplotlib.pyplot as plt
import math
from bidi.algorithm import get_display
from arabic_reshaper import reshape

def d(t):
    return get_display(reshape(t))
```

<div style="float: right; direction: rtl; font-size: 20px"><h2>1. سؤال نخست</h2>&emsp;<span>‫چند‬ ‫درصد‬ ‫از‬ ‫جمعیت‬ ‫‪15‬‬ ‫تا‬ ‫‪64‬‬ ‫سال‬ ‫شاغل‬ ‫هستند‬ ‫؟‬</span></div>

```python
data = p.read_csv("gorgan_census95_activity.csv")
# در اینجا حمعیت بین ۱۵ تا ۶۵ سال را از مجموعه اصلی جمعیت جدا میکنیم
# به این شکل که اجتماع دو مجموعه ۱. اعضایی که سن‌شان نابیشتر از ۶۵ است و
# مجموعه ۲. اعضایی که سنشان ناکمتر از ۱۵ است را در مجموعه هدف قرار می‌دهیم
gp = data[ (data["age_group"] != "15-19") & (data["age_group"] != "65+") ]
gp
```

```python
# در این قسمت از کسر زیر استفاده کرده و مقدار فراوانی اشخاص شاغل را بدست می‌آوریم
# EMPLOYED = ACTIVE / TOTAL
active = gp["p_active_emp"].sum()
total = gp["p_total"].sum()
employed = active / total
```

```python
print(f"{'% EMPLOYED':^50}{'ACTIVE':^50}{'TOTAL':^50}")
print(f"{employed*100:^50.2f}{active:^50}{total:^50}")
```

<div style="text-align: center; direction: rtl; font-size: 20px; margin: 5%; color: darkslategray;">✅&emsp;<span>همانطور که از نتایج بالا می‌توان استنتاج کرد،</span>
    <b>%38.67</b>
از جمعیت بین ۱۵ تا ۶۵ سال شاغل هستند.
</div>

<div style="float: right; direction: rtl; font-size: 20px"><h2>2. سؤال دوم</h2>&emsp;<span>‬درصد جمعیت شاغل 15 تا 64 سال در شهر و روستا چند واحد درصد تفاوت دارد؟</span></div>

```python
# حال، داده ها را بر اساس محل سکونت جدا کرده و 
# مقدایر نظیر ستون ها را در همه سطور جمع می‌زنیم
by_settelment = gp.groupby(["settlement"]).sum(numeric_only=True)
by_settelment
```

```python
# با توجه به اینکه فراوانی جمعیت شهری و روستایی شاغل، از رابطه ذیل حاصل می‌آید
# EMPLOYED = ACTIVE / TOTAL
# فراوانی جمعیت شاغل را در شهر و روستا بطور جداگانه حساب می‌کنیم
ub = by_settelment.iloc[0,0] / by_settelment.iloc[0,7]  # روستايي: p_active_emp / p_total
ru = by_settelment.iloc[1,0] / by_settelment.iloc[1,7]  # شهري: p_active_emp / p_total
# مقدار تفاضل درصد حمعیت شاغل در شهر و روستا:
diff = math.fabs( ub - ru )
print(f"{'% DIFF':^50}{'% UBER':^50}{'% VILLAGR':^50}")
print(f"{diff*100:^50.2f}{ub*100:^50.3f}{ru*100:^50.3f}")
```

<div style="text-align: center; direction: rtl; font-size: 20px; margin: 5%; color: darkslategray;">✅&emsp;<span>با توجه به نتایج بالا می‌توان گفت میزان تفاوت میان درصدهای شاغلان روستا و شهر</span>
    <b>%0.79</b>
است.
</div>

<div style="float: right; direction: rtl; font-size: 20px"><h2>3. سؤال سوم</h2>&emsp;<span>درصد جمعیت شاغل 15 تا 64 سال درمیان زنان و مردان چند واحد درصد تفاوت دارد؟</span></div>

```python
# داده ها را بر اساس جنسیت جدا کرده و 
# مقدایر نظیر ستون ها را در همه سطور جمع می‌زنیم
by_gender = gp.groupby(["gender"]).sum(numeric_only=True)
by_gender
```

```python
# با توجه به اینکه فراوانی جمعیت زن و مرد شاغل، از رابطه ذیل حاصل می‌آید
# {MASC, FEM}_EMPLOYED = {MASC, FEM}_EMP / {MASC, FEM}_TOTAL
# فراوانی جمعیت شاغل زن و مرد رابطور جداگانه حساب می‌کنیم
fem = by_gender.iloc[0,0] / by_gender.iloc[0,7]
masc = by_gender.iloc[1,0] / by_gender.iloc[1,7]
# مقدار تفاضل درصد حمعیت شاغل مرد و زن:
diff_fm = math.fabs( fem - masc )
print(f"{'% DIFF':^50}{'% FEMININE':^50}{'% MASCULINE':^50}")
print(f"{diff_fm*100:^50.2f}{fem*100:^50.1f}{masc*100:^50.1f}")
```

<div style="text-align: center; direction: rtl; font-size: 20px; margin: 5%; color: darkslategray;">✅&emsp;<span>از نتایج بالا می‌توان پی برد میزان تفاوت میان درصد جمعیت شاغل زن و مرد</span>
    <b>%50.24</b>
است.
</div>

<div style="float: right; direction: rtl; font-size: 20px"><h2>4. سؤال چهارم</h2>&emsp;<span>
    چنانچه انتظار میرود درصد جمعیت شاغل با افزایش سن افزایش می‌یابد، در یکی ازگروههای سنی به
اوج میرسد و سپس کاهش می‌یابد.

 - درصد جمعیت شاغل در کدام گروههای سنی به اوج خود میرسد؟
 - در چه سنی کاهش چمشگیر درصد جمعیت شاغل آغاز میشود؟
    </span></div>

```python
# داده ها را بر اساس گروه سنی جدا کرده و 
# مقدایر نظیر ستون ها را در همه سطور جمع می‌زنیم
by_age = gp.groupby(["age_group"]).sum(numeric_only=True)
by_age
```

```python
# جمعیت را بر اساس گروه سنی جدا کرده و سپس با توجه به وضعیت اشتغال ایشان
# آنها را دسته بندی کرده نمایش می‌دهیم.
plt.bar(by_age.index, by_age['p_active_emp'], color='orange')
plt.plot(by_age.index, by_age['p_active_emp'], color='yellow')
plt.title(d("نمودار وضعیت اشتغال بر حسب گروه سنی"))
plt.xticks(rotation=60)
plt.yticks(rotation=60)
plt.xlabel(d("گروه سنی"))
plt.ylabel(d("وضعیت اشتغال"))
```

<div style="text-align: center; direction: rtl; font-size: 20px; margin: 5%; color: darkslategray;">✅&emsp;<span>نتایج بدست آمده با نیم‌نگاهی به نمودار فوق:</span>
    
1.    <b>درصد جمعیت شاغل در سنین ۲۵-۴۴ در بیشترین حالت خود است.</b>
2.    <b>در سین ۴۹-۵۰ سالگی، مقدار درصد جمعیت شاغل با افت چشمگیر مواجه است.</b>
</div>

<div style="float: right; direction: rtl; font-size: 20px"><h2>5. سؤال پنجم</h2>&emsp;<span>
شاغلان روستا زودتر دست از کار می‌کشند یا شاغلان شهر؟</span></div>

```python
# شاغلان را بر اساس سن بازنشسنگی و محل سکونت دسته بندی میکنیم:
people = dict()
colors = (("blue", "orange"), ("red", "lightgreen"))
for i in ['شاغلان شهري', 'شاغلان روستايي']:
    s = 0 if i != 'شاغلان شهري' else 12
    f = s*2 if s == 12 else 12
    # فراوانی جمعیت شاغل را تسبت به کل جمعیت بدست می‌آوریم:
    t = data.groupby(["settlement","age_group"]).sum(numeric_only=True)[s:f]
    people[i] = t['p_active_emp'] / t['p_total']*100

# نمودار مربوطه را رسم می‌کنیم:
for i in range(1, 3):
    k = list(people.keys())[i-1]
    plt.subplot(1, 2, i)
    plt.bar(data["age_group"][:12], people[k], color=colors[i-1][0])
    plt.plot(data["age_group"][:12], people[k], color=colors[i-1][1])
    plt.xlabel(d("گروه سنی"))
    plt.ylabel(d(k), labelpad=-225 if i == 2 else 10)
    plt.xticks(rotation=60)
    plt.yticks(rotation=60)

plt.title(d("نمودار مقدار فعالیت شاغلان روستا و شاغلان شهر بر اساس گروه سنی آنها"),
          pad='50', loc='right')
```

<div style="text-align: center; direction: rtl; font-size: 20px; margin: 5%; color: darkslategray;">✅&emsp;<span>بدست می‌آید که </span>
    <b>شاغلان شهری</b>
زودتر دست از کار می‌کشند و همچنین شیب بازنشسته شدن در شهر بیشتر از روستا‌ست.
</div>

<div style="float: right; direction: rtl; font-size: 20px"><h2>6. سؤال ششم</h2>&emsp;<span>
آیا الگوی متفاوت در سن دست از کار کشیدن جمعیت شهری و روستایی، در میان زنان و مردان متمایز
است؟</span></div>

```python
empl = dict()
colors = [["pink", "yellow"], ["cyan", "orange"], ["green", "blue"], ["yellow", "red"]]

# ابتدا داده‌ها را دسته بندی می‌کنیم: مرد و زن روستایی -- مرد و زن شهری
# _*_|__شهری__|__روستایی__|
# مرد|      +    |    +   |
# زن |      +    |    +   |
for i in ["شهري", "روستايي"]:
    for j in ["زن", "مرد"]:
        if i not in empl:
            empl[i] = {}
        empl[i][j] = data[(data['gender'] == j) & (data["settlement"] == i)]

for i, j in zip(range(1, 5), range(4, 0, -1)):
    iindex = i % 2
    jindex = (i // 2) % 2
    sett = list(empl.keys())[iindex]
    gen = list(empl[sett].keys())[jindex]
    t = empl[sett][gen]['p_active_emp']
    plt.subplot(2, 2, i)
    plt.plot(data["age_group"][:12], t, color=colors[i-1][0])
    plt.bar(data["age_group"][:12], t, color=colors[i-1][1])
    plt.xticks(rotation=60)
    plt.xlabel(d("گروه سنی"), labelpad=10)
    plt.ylabel(d(f"{sett} - {gen}"), labelpad= 15 if (i%2==1) else -220)
    plt.xticks(rotation=60)

plt.title(d("نمودار الگوی دست از کار کشیدن کارمندان زن و کارمندان مرد در شهر و روستا"),
         pad='10', loc='right')
```
